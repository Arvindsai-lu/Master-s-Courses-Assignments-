```python
  # 3. Perform PCA on the data. Remember to center the data points first by computing the mean
# data vector µ and subtracting it from every point. With the centered data matrix, do an SVD and
# compute the principal components.

import numpy as np

# Step 1: Center the data by computing the mean data vector µ
mean_data = np.mean(X, axis=0)

# Center the data points by subtracting the mean vector from every point
centered_data = X - mean_data

# Step 2: Perform Singular Value Decomposition (SVD)
U, S, Vt = np.linalg.svd(centered_data, full_matrices=False)


# Step 3: Compute the principal components
principal_components = U

# To Print the first two principal components (v1 and v2)
v1 = principal_components[:, 0]
v2 = principal_components[:, 1]

# Let's print the first two principal components
print("Principal Component 1 (v1):")
print(v1)
print("\nPrincipal Component 2 (v2):")
print(v2)

#finally we done with these coding
```
